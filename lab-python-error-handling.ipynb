def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))

                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True

````python
from typing import Dict, List


def initialize_inventory(products: List[str]) -> Dict[str, int]:
    """
    Initializes the inventory dictionary by prompting the user for the quantity
    of each product in the given list.

    Args:
        products: A list of product names.

    Returns:
        A dictionary representing the inventory, where keys are product names
        and values are their corresponding quantities (integers).
    """
    inventory: Dict[str, int] = {}
    print("\nPlease enter the quantity of the following products:")
    for item in products:
        while True:  # Loop until valid input is provided
            quantity_str = input(f"{item}: ")
            try:
                quantity = int(quantity_str)
                if quantity >= 0:  # Ensure the quantity is a non-negative number
                    inventory[item] = quantity
                    break  # Exit the inner loop if input is valid
                else:
                    print("Please enter a non-negative number for the quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid integer.")
    return inventory


def get_customer_orders(inventory: Dict[str, int]) -> List[Dict[str, int]]:
    """
    Prompts the user for the number of orders and the products for each order.

    Args:
        inventory: A dictionary representing the current inventory. Keys are product
                   names (strings), and values are the quantities available (integers).

    Returns:
        A list of dictionaries, where each dictionary represents an order.
        Each order dictionary contains product names as keys and the quantity
        ordered as values.
    """

    orders: List[Dict[str, int]] = []  # Creating an empty list to store orders

    while True:
        try:
            num_orders = int(input('Enter the number of orders: '))
            if num_orders > 0:
                break
            else:
                print('Number of orders must be a positive integer. Please try again.')
        except ValueError:
            print('Invalid input. Please enter a valid integer for the number of orders.')

    for _ in range(num_orders):
        order: Dict[str, int] = {}  # Initializing an empty dictionary for the current order
        while True:
            product_name = input('Enter the product name (or "done" to finish order): ').lower()  # Convert input to lowercase for case-insensitive matching

            if product_name == 'done':
                break  # Move to the next order

            if product_name not in inventory:
                print(f"Error: Product '{product_name}' is not in the inventory. Please enter a valid product name.")
                continue  # restart current iteration of the loop

            if inventory[product_name] <= 0:
                print(f"Error: Product '{product_name}' is out of stock. Please enter a different product.")
                continue  # restart current iteration of the loop

            while True:
                try:
                    quantity = int(input(f"Enter the quantity of {product_name} to order: "))
                    if quantity > 0 and quantity <= inventory[product_name]:
                        order[product_name] = quantity
                        break
                    elif quantity <= 0:
                        print("Quantity must be positive")
                    else:
                        print(f"Error: Not enough stock available for '{product_name}'.  Only {inventory[product_name]} available.")
                except ValueError:
                    print('Invalid input. Please enter a valid integer for the quantity.')

        orders.append(order)  # Add the current order to the list of orders

    return orders


def calculate_total_price(orders: List[Dict[str, int]], price_list: Dict[str, float]) -> float:
    """
    Calculates the total price of all orders, using a price list.

    Args:
        orders: A list of order dictionaries.
        price_list: A dictionary of prices for each product.

    Returns:
        The total price of all orders.  Returns -1.0 if there's a pricing error.
    """
    total_price: float = 0.0

    for order in orders:
        for product, quantity in order.items():
            if product not in price_list:
                print(f"Error: Price for product '{product}' not found.  Cannot calculate total price.")
                return -1.0  # Indicate an error
            total_price += price_list[product] * quantity

    return total_price



product_list = ["t-shirt", "mug", "hat", "book"]
inventory = initialize_inventory(product_list)  # Example Inventory
price_list = {"t-shirt": 15.0, "mug": 8.0, "hat": 10.0, "book": 20.0}
customer_orders = get_customer_orders(inventory)

print("Customer Orders:")
for i, order in enumerate(customer_orders):
    print(f"Order {i + 1}: {order}")

total = calculate_total_price(customer_orders, price_list)
if total != -1.0:
    print(f"Total price for all orders: ${total:.2f}")
else:
    print("Could not calculate the total price due to an error.")


# Testing the error handling
print("\n--- Testing Error Handling ---")
# 1. Invalid Number of Orders (Non-integer)
print("\nTest 1: Invalid Number of Orders (Non-integer)")
# (The program will prompt you to re-enter)
# Enter "abc" when prompted for the number of orders.

# 2. Invalid Number of Orders (Negative)
print("\nTest 2: Invalid Number of Orders (Negative)")
# (The program will prompt you to re-enter)
# Enter "-1" when prompted for the number of orders.

# 3. Invalid Product Name
print("\nTest 3: Invalid Product Name")
# Enter "invalid_product" as a product name.

# 4. Out of Stock Product
print("\nTest 4: Out of Stock Product")
#  If you initialized "book" to 0, try to order it.

# 5. Invalid Quantity (Non-integer)
print("\nTest 5: Invalid Quantity (Non-integer)")
# Enter "abc" when prompted for the quantity.

# 6. Invalid Quantity (Negative)
print("\nTest 6: Invalid Quantity (Negative)")
# Enter "-1" when prompted for the quantity.

# 7. Invalid Quantity (Too Large)
print("\nTest 7: Invalid Quantity (Too Large)")
# Enter a quantity larger than the available stock.

# 8. Product with missing Price
print("\nTest 8: Product with missing Price")
# Manually add an order for "unknown_item" (you'll have to edit customer_orders directly)
# price_list doesn't have it
# Remove or comment this block after the testing is done, as it requires to comment a line
customer_orders.append({"unknown_item": 1})  # Add an order with an unknown item
total = calculate_total_price(customer_orders, price_list)
if total != -1.0:
    print(f"Total price for all orders: ${total:.2f}")
else:
    print("Could not calculate the total price due to an error.")
````
